# Data Visualization

Data visualization refers to the graphical depiction of information and data. Data visualization tools use visual components such as charts, graphs, and maps to make trends, outliers, and patterns in data more visible and understandable.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Basic of Visualization

### Using Matplotlib

#### 1. Comparison - Line Chart

```python
beam = [50, 39, 45, 57, 44] #beam usage per day
```

after we define the data points of beam, we are going to plot its line chart using Matplotlib

```python
plt.plot(beam)
```

to better understand the graph and its purpose, we can add the x-axis values too

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
beam = [50, 39, 45, 57, 44]
```

since we have the 'days' variable as the x-axis, we need to type it before the beam variable below

```python
plt.plot(days, beam) #plt.plot(x, y)
```

let's add labels to the axes so that we can show what each acis represents

```python
plt.plot(days, beam)
plt.xlabel('Day')
plt.ylabel('Beam (usage per day)')
```

#### 2. Comparison - Multiple Line Chart

to plot multiple datasets on the same graph, just use the plt.plot function once for each dataset. let's use this to compare the beam vs odong as public transportation in Padjadjaran University on the same graph

```python
beam = [50, 39, 45, 57, 44]
odong = [78, 90, 84, 96, 68]
```

```python
plt.plot(days, beam)
plt.plot(days, odong)
plt.xlabel('Day')
plt.ylabel('Public Transportation (usage per day)');
```

we can add a legend which tells us what each line in our graph means. to understand what we are plotting, we can add a title to our graph

```python
plt.plot(days, beam)
plt.plot(days, odong)

plt.xlabel('Day')
plt.ylabel('Transportation (per student)')

plt.title("Use of Public Transportation in Padjadjaran University")
plt.legend(['Beam', 'Odong'])
```

#### 3. Comparison - Bar Chart

```python
plt.bar(days, beam) #plt.bar(x, y)

plt.xlabel('Day')
plt.ylabel('Transportation (per student)')

plt.title("Use of Beam as Public Transportation in Padjadjaran University")
```

```python
#stacked bar

plt.bar(days, beam)
plt.bar(days, odong, bottom=beam)
```

#### 2. Comparison - Bar Chart

Lets use the tips dataset in Seaborn next.

The data contains passenger trip records with details:
- pick-up timestamp and destination timestamp
- number of passengers
- drop-off distance
- taxi fare
- tip amount
- toll fee
- taxi type by colour
- payment type
- location of pick-up area and destination.

this **Taxis Dataset** is a dataset provided by seaborn library, so you can directly load the dataset

```python
taxis_df = sns.load_dataset("taxis")
taxis_df.head()
```

```python
taxis_df.info()
```

we can draw a bar chart to visualize how the average fare amount varies accross different taxi type by colour.

```python
sns.barplot(x='color',
            y='fare',
            data=taxis_df)
```

if you want to compare bar plots side-by-side, you can use the hue argument. the comparison will be done based on the third feature specified in this argument

```python
sns.barplot(x='color', y='fare', hue='payment', data = taxis_df)
```

```python
#add a title
plt.title("Distribution of Taxis Transaction")

sns.barplot(x='color', y='fare', hue='payment', data = taxis_df)
```

you can make the bars horizontal by switching the axes

```python
sns.barplot(x='fare', y='color', hue='payment', data = taxis_df)
```

```python
#add title
plt.title("Distribution of Taxis Transaction")

sns.barplot(x='fare', y='color', hue='payment', data = taxis_df)
```

#### Using Matplotlib

```python
import numpy as np

#specifying the number of bins
plt.hist(healthexp_df.Life_Expectancy, bins=np.arange(70, 87, 0.75));
```
